```python
import torch
import numpy as np
import onnx
import os
import WDNet
import matplotlib.pyplot as plt
import torchvision
import PIL
```

```python
device = 'cpu'
gen_model = WDNet.generator(3,3)
gen_model.load_state_dict(torch.load(os.path.join('../model_WDNet','WDNet_G.pkl')))
gen_model.eval()
gen_model.to(device)
```

```python
#pick-up test image
test_img_path = os.path.join('../model_WDNet','my_test_img','wm1.jpg')
#test_img_path = os.path.join('../data_CLWD/test/Watermarked_image/18.jpg')
test_img = torchvision.io.read_image(test_img_path)
print(f'input shape: {test_img.shape}')
test_img = test_img.float() / 255
test_img = test_img.unsqueeze(0)
test_img = test_img.to(device)

with torch.no_grad():
    restored_img = gen_model(test_img)
```

```python
import torchvision
torchvision.utils.save_image(restored_img[0], '../restored.png')
pil_image = torchvision.transforms.functional.to_pil_image(restored_img[0].squeeze(0))
pil_image.show()
```
